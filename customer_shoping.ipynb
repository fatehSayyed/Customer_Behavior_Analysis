# In above we see the there is 37 null values in the "Review Rating" so we replace the null value by there categpry "median".#

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

# convert all the column name in "samll letters" and sapce with "_" .

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
df.columns
```

# replace the column name on purcahse_amount_(usd) with "purchase_amount" #

```python
df.rename(columns={"purchase_amount_(usd)": "purchase_amount"}, inplace=True)
```

```python
df.columns
```

# craete the column age group # 

```python
labels = ['Youn-adult' , 'Adult' ,'Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df.head(10)
```

# create the column "purchase_frequency_days"

```python
frequency_mapping = {
    "Weekly": 7,
    "Bi-Weekly": 14,
    "Monthly": 30,
    "Quarterly": 90,
    "Annually": 365,
    'Every 3 month': 90,
    'Fortnightly': 14,
    'Every 3 Months': 90
}

df["frequency_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```
